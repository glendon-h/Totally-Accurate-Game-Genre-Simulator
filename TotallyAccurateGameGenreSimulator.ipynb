```python


gameList = ["Dark Souls",
            "Zelda",
            "Metroid",
            "Grand Theft Auto",
            "Portal",
            "Minecraft",
            'Elder Scrolls',
            'Batman',
            'Super Meat Boy',
            'Super Mario',
            'StarCraft',
            'Red Dead Redemption',
            'Mass Effect',
            'Uncharted',
            'Left 4 Dead',
            'Assassin`s Creed',
            'Spelunky',
            'Persona',
            'Fallout',
            'Team Fortress',
            'Halo',
            'BioShock',
            'Call of Duty',
            'Hitman',
            'Gears of War',
            'Guitar Hero',
            'God of War',
            'Civilization',
            'World of Warcraft',
            'Metal Gear',
            'Half-Life',
            'Burnout',
            'Super Smash Bros.',
            'Final Fantasy',
            'The Sims',
            'Diablo',
            'Baldur`s Gate',
            'Shenmue',
            'Quake',
            'Xcom',
            'Myst',
            'Doom',
            'Rogue']

gameSuffixes = ['vania','like','esque','lite','borne']

gamePrefixes = ['Stealth',
                'Text',
                'Graphic',
                'Visual',
                'Interactive',
                'Real-time',
                'Action',
                'Tactical',
                'Battle',
                'Fantasy',
                'First-Person',
                'Third-Person',
                'Top-Down',
                'Turn-Based',
                'Grand Strategy',
                'Competitive',
                'Casual',
                'Party',
                'Trivia',
                'Idle',
                'Art',
                'Hardcore',
                'Permadeath',
                'Horror',
                'Sandbox',
                'Party-based',
                'Tower Defense',
                'Racing',
                'Mobile',
                'Logic',
                'VR',
                'AR',
                'Physical',
                'Crunchy']

gameGenres = ['MMO',
              'Platformer',
              'Simulator',
              'Adventure',
              'Puzzle',
              'Action',
              'Shooter',
              'Combat',
              'Sports',
              'Role Playing',
              'Educational',
              'Fighting',
              'Survival',
              'Rhythm',
              'Horror',
              'Movie',
              'Novel',
              'Management',
              'Wargame',
              'Royale',
              'Boardgame',
              'Cardgame']

gameList_len = len(gameList)
gamePrefixes_len = len(gamePrefixes)
gameSuffixes_len = len(gameSuffixes)
gameGenres_len = len(gameGenres)

def getGameGenre(counter):
    
    arg = int(counter)
    
    for i in range(arg):
        gameList_state = random.random()
        gameList_i = random.randint(0, gameList_len - 1)
        gameSuffixes_i = random.randint(0, gameSuffixes_len - 1)
        gamePrefix_state = random.random()
        gamePrefix_i = random.randint(0, gamePrefixes_len - 1)
        gamePrefix_j = random.randint(0, gamePrefixes_len - 1)
        gamePrefix_k = random.randint(0, gamePrefixes_len - 1)

        while gamePrefix_j == gamePrefix_i:
            gamePrefix_j = random.randint(0, gamePrefixes_len - 1)

        while gamePrefix_k == gamePrefix_i or gamePrefix_k == gamePrefix_j:
            gamePrefix_k = random.randint(0, gamePrefixes_len - 1)

        gameGenre_state = random.random()
        gameGenre_i = random.randint(0, gameGenres_len - 1)
        gameGenre_j = random.randint(0, gameGenres_len - 1)
        gameGenre_k = random.randint(0, gameGenres_len - 1)

        while gameGenre_j == gameGenre_i:
            gameGenre_j = random.randint(0, gameGenres_len - 1)

        while gameGenre_k == gameGenre_i or gameGenre_k == gameGenre_j:
            gameGenre_k = random.randint(0, gameGenres_len - 1)

        gameGenreString = gameList[gameList_i]

        if gameList_state > 0.3:
            gameGenreString = gameGenreString + gameSuffixes[gameSuffixes_i]

        if gamePrefix_state < 0.6:
            gameGenreString = gameGenreString + " " + gamePrefixes[gamePrefix_i]

        elif 0.6 < gamePrefix_state < 0.9:
            gameGenreString = gameGenreString + " " + gamePrefixes[gamePrefix_i] + " " + gamePrefixes[gamePrefix_j]

        elif gamePrefix_state > 0.9:
            gameGenreString = gameGenreString + " " + gamePrefixes[gamePrefix_i] + " " + gamePrefixes[gamePrefix_j] + " " + gamePrefixes[gamePrefix_k]

        if gameGenre_state < 0.6:
            gameGenreString = gameGenreString + " " + gameGenres[gameGenre_i]

        elif 0.6 < gameGenre_state < 0.9:
            gameGenreString = gameGenreString + " " + gameGenres[gameGenre_i] + " " + gameGenres[gameGenre_j]

        elif gameGenre_state > 0.9:
            gameGenreString = gameGenreString + " " + gameGenres[gameGenre_i] + " " + gameGenres[gameGenre_j] + " " + gameGenres[gameGenre_k]

        print(gameGenreString)

button = widgets.Button(
    description='Generate',
)
@button.on_click
def generate_on_click(b):
    getGameGenre(1)
button
```
